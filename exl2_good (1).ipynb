نموذج **ExLlamaV2 (exl2)** مصمم ليكون أكثر كفاءة من النماذج التقليدية مثل `HF Transformers`، ويستهلك ذاكرة VRAM أقل بسبب ضغط الوزن وتقنيات التحميل الفعالة.  

### **متطلبات الذاكرة (VRAM) حسب حجم النموذج ودرجة الضغط:**
#### **Llama 3.1 - 8B Instruct (exl2)**
| **درجة الضغط** | **استهلاك VRAM** (تقريبي) |
|--------------|----------------|
| 4-bit (exl2) | ~7-9 GB |
| 3-bit (exl2) | ~5-7 GB |
| 2-bit (exl2) | ~4-5 GB |

#### **Llama 3.1 - 70B (exl2)**
| **درجة الضغط** | **استهلاك VRAM** (تقريبي) |
|--------------|----------------|
| 4-bit (exl2) | ~35-40 GB |
| 3-bit (exl2) | ~24-30 GB |
| 2-bit (exl2) | ~18-22 GB |

### **كيف تتأكد من استهلاك النموذج للـ VRAM؟**
إذا كنت تريد معرفة استهلاك النموذج للذاكرة أثناء التشغيل، يمكنك استخدام:
```python
import torch

# عرض استهلاك الذاكرة على الـ GPU
print(f"VRAM Used: {torch.cuda.memory_allocated() / 1e9:.2f} GB")
print(f"VRAM Reserved: {torch.cuda.memory_reserved() / 1e9:.2f} GB")
```

### **أفضل GPU لتشغيل ExLlamaV2**
- **8GB VRAM** → يمكنك تشغيل `8B (exl2 4-bit)` لكن بدون مساحة كافية للسياق الطويل.
- **12GB VRAM (مثل RTX 3060)** → جيد لـ `8B (exl2 4-bit)` مع أداء معقول.
- **24GB VRAM (مثل RTX 4090 أو A6000)** → يمكن تشغيل `8B (exl2 3-bit)` أو `70B (exl2 4-bit)`.
- **48GB+ VRAM (A100, H100, MI300X)** → الأفضل للنماذج الكبيرة جدًا مثل `70B (exl2 3-bit)`.

### **الاستنتاج**
- **8B (exl2 4-bit) يحتاج 7-9GB VRAM**.
- إذا كان لديك **RTX 3060 (12GB)** أو أعلى، يمكنك تشغيله بسلاسة.
- كلما كانت **درجة الضغط أقل** (مثل 3-bit أو 2-bit)، كان استهلاك الذاكرة أقل لكنه قد يؤثر على الدقة.

هل لديك GPU معين تريد التأكد من إمكانية تشغيل النموذج عليه؟ 🚀
